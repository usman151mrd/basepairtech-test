```python
import matplotlib.pyplot as plt
import json
```

```python
new_router = dict()
file = open("new_router.json","r")
new_router = json.load(file)
file.close()
print(new_router)
```

```python
old_router = dict()
file = open("old_router.json","r")
old_router = json.load(file)
file.close()
print(old_router)
```

```python
fig = plt.figure(figsize=(15, 10), dpi=300)
# plt.axes([0,25,0,25])
plt.subplot(2, 1, 1)
plt.barh(list(new_router.keys()),list(new_router.values()))
plt.title("New Router with designation : "+new_designation)
plt.subplot(2, 1, 2)
plt.barh(list(old_router.keys()), list(old_router.values()))
plt.title("Old Router with designation : "+old_designation)

plt.suptitle("Old vs New Router")
plt.show()
fig.savefig("routers.png", format="png", orientation='landscape', facecolor='auto', edgecolor='auto')
```
